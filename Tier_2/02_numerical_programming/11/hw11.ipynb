### Task - Backpack problem

#### 1. Download the PyGad library


```python
import pygad
import numpy as np
import matplotlib.pyplot as plt
```

Input data:


```python
class Product:
    def __init__(self, name, space, price):
        self.name = name
        self.space = space
        self.price = price


products_list = []
products_list.append(Product("Refrigerator A", 0.751, 999.90))
products_list.append(Product("Cell phone", 0.00000899, 2199.12))
products_list.append(Product("TV 55", 0.400, 4346.99))
products_list.append(Product("TV 50' ", 0.290, 3999.90))
products_list.append(Product("TV 42' ", 0.200, 2999.00))
products_list.append(Product("Notebook A", 0.00350, 2499.90))
products_list.append(Product("Ventilator", 0.496, 199.90))
products_list.append(Product("Microwave A", 0.0424, 308.66))
products_list.append(Product("Microwave B", 0.0544, 429.90))
products_list.append(Product("Microwave C", 0.0319, 299.29))
products_list.append(Product("Refrigerator B", 0.635, 849.00))
products_list.append(Product("Refrigerator C", 0.870, 1199.89))
products_list.append(Product("Notebook B", 0.498, 1999.90))
products_list.append(Product("Notebook C", 0.527, 3999.00))

space_limit = 3
```

#### 2. Define the fitness function separately in the code


```python
def fitness(self, solution, solution_idx) -> float:
    score = 0
    sum_spaces = 0

    for i in range(len(solution)):
        score += products_list[i].price * solution[i]
        sum_spaces += products_list[i].space * solution[i]

    if sum_spaces > space_limit:
        score = 1

    return score
```

#### 3. Determine the initial population


```python
number_genes = len(products_list)


def get_population(num_genes, pop_size=10):
    return np.random.randint(2, size=(pop_size, num_genes))


initial_population = get_population(number_genes)
initial_population
```

#### 4. Set the parameters of the genetic algorithm: the number of populations, the number of parental chromosomes participating in the crossover, the type of crossover and mutation, etc.


```python
pop_size = 10
num_generations = 100
num_parents_mating = 4
crossover_type = ["single_point", "two_points", "uniform", "scattered"]
mutation_type = ["random", "swap", "inversion", "scramble"]
mutation_percent_genes = 10
```

#### 5. Create an instance of a generic algorithm using PyGad


```python
model = pygad.GA(
    num_generations=num_generations,
    num_parents_mating=num_parents_mating,
    fitness_func=fitness,
    sol_per_pop=pop_size,
    num_genes=number_genes,
    initial_population=initial_population,
    parent_selection_type="sss",
    mutation_percent_genes=mutation_percent_genes,
)
```

#### 6. Investigate the effect of various crossovers and mutations on the result


```python
for crossover in crossover_type:
    for mutation in mutation_type:
        fitness_over_time = []

        def callback_generation(model):
            fitness_over_time.append(model.best_solution()[1])

        model = pygad.GA(
            num_generations=num_generations,
            num_parents_mating=num_parents_mating,
            fitness_func=fitness,
            sol_per_pop=pop_size,
            num_genes=number_genes,
            initial_population=initial_population,
            parent_selection_type="sss",
            crossover_type=crossover,
            mutation_type=mutation,
            mutation_percent_genes=mutation_percent_genes,
            on_generation=callback_generation,
        )

        model.run()

        plt.plot(fitness_over_time)
        plt.title(f"Crossover: {crossover}, Mutation: {mutation}")
        plt.xlabel("Generation")
        plt.ylabel("Fitness")
        plt.show()

        solution, solution_fitness, _ = model.best_solution()
        total_volume = np.sum(
            np.array([products_list[i].space for i in range(len(solution))]) * solution
        )

        print(f"Crossover: {crossover}, Mutation: {mutation}")
        print(f"Best solution: {solution}")
        print(f"Total price: {solution_fitness}")
        print(f"Total volume: {total_volume}")
        print("-" * 30)
```

#### 7. To draw a conclusion about the effectiveness of the application of the genetic algorithm and the optimal set of its parameters


As a result of the research, the combination "Crossover: uniform, Mutation: inversion" coped best with the task. She is not the only one who found the most value in the backpack problem, but also the fastest to converge.
