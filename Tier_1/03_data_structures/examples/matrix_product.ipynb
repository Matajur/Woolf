```python
import matplotlib.pyplot as plt
import numpy as np


def multiply_matrices(matrix_1, matrix_2):
    matrix_dim = len(matrix_1)
    result = [[0 for _ in range(matrix_dim)] for _ in range(matrix_dim)]

    for i in range(matrix_dim):
        for j in range(matrix_dim):
            for k in range(matrix_dim):
                result[i][j] += matrix_1[i][k] * matrix_2[k][j]

    return result


A = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
B = [[10, 11, 12], [13, 14, 15], [16, 17, 18]]

print(multiply_matrices(A, B))

n_range = np.arange(1, 100)

# Calculation of values for different time complexity
unit_vector = np.ones_like(n_range)
linear_complexity = n_range
squared_complexity = n_range**2
cubed_complexity = n_range**3

plt.figure(figsize=(12, 8))
plt.plot(n_range, unit_vector, label="O(1)")
plt.plot(n_range, linear_complexity, label="O(n)")
plt.plot(n_range, squared_complexity, label="O(n^2)")
plt.plot(n_range, cubed_complexity, label="O(n^3)")
plt.yscale("log")  # Logarithmic scale for the y-axis
plt.xlabel("n")
plt.ylabel("Operations")
plt.title("The main time complexity of algorithms")
plt.legend()
plt.grid(True, which="both", ls="--", c="0.65")
plt.show()
```
