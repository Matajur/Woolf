### Task 1
Solve the problem with a text condition analytically

3 types of specialists worked on the project - designers, developers and testers. We know the number of hours in each stage of development and the amount of money that is invested in the execution of each stage as a cost.
The manager divided the entire development into 3 stages: creating a design, developing the software part, and final testing.

At the first stage, 20 hours were worked by the designer and 3 hours by the developers - at this stage, 775 dollars were allocated to the estimate.
At the second stage, designers worked 8 hours, developers - 28, and testers - 4. This stage was estimated at $1012.
At the final stage, designers worked 4 hours, developers - 12, and testers 32. Labor resources at this stage are estimated at $696.

Find the hourly cost of each specialist by solving a system of linear equations based on data by development stage.

Resolution:

$$
\overline x =
\left (\begin{array}{cc}
x - ? \\
y - ? \\
z - ?
\end{array}\right)
\\
A =
\left (\begin{array}{cc}
20 & 3 & 0 \\
8 & 28 & 4 \\
4 & 12 & 32
\end{array}\right)
\\
b =
\left (\begin{array}{cc}
775 \\
1012 \\
696
\end{array}\right)
\\
A \overline x = b$$

$\Delta = \left|A\right| = \left|\begin{array}{cc}
20 & 3 & 0 \\
8 & 28 & 4 \\
4 & 12 & 32
\end{array}\right| = 20 \cdot 28 \cdot 32 + 3 \cdot 4 \cdot 4 + 0 \cdot 8 \cdot 12 - 4 \cdot 28 \cdot 0 - 8 \cdot 3 \cdot 32 - 20 \cdot 12 \cdot 4 =\\
= 17.920 + 48 + 0 - 0 - 768 - 960 = 16.240$

Cost of designers:

$\Delta x = \left|\begin{array}{cc}
775 & 3 & 0 \\
1012 & 28 & 4 \\
696 & 12 & 32
\end{array}\right| = 775 \cdot 28 \cdot 32 + 3 \cdot 4 \cdot 696 + 0 \cdot 1012 \cdot 12 - 696 \cdot 28 \cdot 0 - 1012 \cdot 3 \cdot 32 - 775 \cdot 12 \cdot 4 =\\
= 694.400 + 8.352 + 0 - 0 - 97.152 - 37.200 = 568.400$

$$x = \frac{\Delta x}{\Delta} = \frac{568.400}{16.240} = 35$$

Cost of developers:

$\Delta y = \left|\begin{array}{cc}
20 & 775 & 0 \\
8 & 1012 & 4 \\
4 & 696 & 32
\end{array}\right| = 20 \cdot 1012 \cdot 32 + 775 \cdot 4 \cdot 4 + 0 \cdot 8 \cdot 696 - 4 \cdot 1012 \cdot 0 - 8 \cdot 775 \cdot 32 - 20 \cdot 696 \cdot 4 =\\
= 647.680 + 12.400 + 0 - 0 - 198.400 - 55.680 = 406.000$

$$y = \frac{\Delta y}{\Delta} = \frac{406.000}{16.240} = 25$$

Cost of QA:

$\Delta z = \left|\begin{array}{cc}

$$z = \frac{\Delta z}{\Delta} = \frac{129.920}{16.240} = 8$$

Verification: $x+y+z = 35 + 25 + 8 = 68$

```python
import numpy as np
A = np.array([[-1, 1, 2], [0, -1, -3], [4, -3, 2]])
B = np.array([1, -4, 7])
```

- Write a function to solve the system using the matrix method
- Write a function to solve the system using Cramer's method

```python
# Function for solving the system by the matrix method
def solve_inv_matrix(a, b, verbose=False):
    if len(a) != len(a[0]) and len(a) != len(b):
        return "Wrong matrix shape"
    det = np.linalg.det(a)
    if not det:
        return "Sungular matrix"
    return np.linalg.inv(a) @ b

print(f"Solution vector: \r\n {solve_inv_matrix(A, B)}")
```

```python
from copy import deepcopy
# A function for solving the system by Cramer's method
def solve_cramer(a, b, verbose=False):
    if len(a) != len(a[0]) and len(a) != len(b):
        return "Wrong matrix shape"
    det = np.linalg.det(a)
    if not det:
        return "Sungular matrix"
    result = []
    for i in range(len(a)):
        var = deepcopy(a)
        for j in range(len(a)):
            var[j,i] = b[j]
        result.append(round(np.linalg.det(var) / det, 5))
    return np.array(result)

print(f"Solution vector: \r\n {solve_cramer(A, B)}")
```
