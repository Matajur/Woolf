## Task 2
Out of 40 programmers, 18 are proficient in Python, 19 in C++, and 21 in Java. It is known that 10 programmers know both Python and C++, 7 - Python and Java, 8 - C++ and Java. Three programmers do not know any of Python, C++, or Java languages. Find the number of programmers who know all three programming languages at the same time.

$Q = P + C + J - PC - PJ - CJ + PCJ + NA$,\
where $Q$ - the total number of programmers,\
$P$ - the number of programmers who know Python,\
$C$ - the number of programmers who know C++,\
$J$ - the number of programmers who know Java,\
$PC$ - the number of programmers who know Python and C++,\
$PJ$ - the number of programmers who know Python and Java,\
$CJ$ - the number of programmers who know C++ and Java,\
$PCJ$ - the number of trilingual programmers,\
$NA$ - the number of programmers who do not speak any of the specified languages.

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn3

total_set = 40
python_set = 18
cpp_set = 19
java_set = 21
illiterate_set = 3
python_cpp = 10
python_java = 7
cpp_java = 8

python_cpp_java = total_set - (python_set + cpp_set + java_set) + (python_cpp + python_java + cpp_java) - illiterate_set
python_only = python_set - python_cpp - python_java + python_cpp_java
cpp_only = cpp_set - python_cpp - cpp_java + python_cpp_java
java_only = java_set - python_java - cpp_java + python_cpp_java
python_cpp_only = python_cpp - python_cpp_java
python_java_only = python_java - python_cpp_java
cpp_java_only = cpp_java - python_cpp_java

venn3(subsets=(python_only, cpp_only, python_cpp_only, java_only,
               python_java_only, cpp_java_only, python_cpp_java),
      set_labels=('Python', 'C++', 'Java'))

plt.title("Venn Diagram of Programming Language Proficiency")
plt.show()
```
