### Task 1

In a batch of 50 parts, 5 are non-standard. Determine the probability that among six parts selected at random for inspection, two will be substandard. Solve analytically.

---
Resolution:



The order of detection of non-standard parts is not important, so this problem belongs to the group of combinations.

Let's enter the following notation: \
$P(A)$ — the probability of extracting two non-standard parts from six,\
where $A$ — the event itself, resulting in 2 non-standard and 4 standard parts,\
$N$ — the total number of parts $(N = 50)$,\
$n$ — the number of selected details $(n = 6)$,\
$K$ — the total number of non-standard parts $(K = 5)$,\
$k$ — the number of non-standard parts among the selected $(k = 2)$.

The total number of combinations that can be used to select $n=6$ parts from $N=50$:
$$C_N^n = \frac{N!}{n! \cdot (N-n)!}\\
C_{50}^6 = \frac{50!}{6! \cdot (50-6)!}$$

Let's count the number of combinations that can be used to choose $2$ non-standard parts from $5$ non-standard parts:
$$C_K^k = \frac{K!}{k! \cdot (K-k)!}\\
C_{5}^2 = \frac{5!}{2! \cdot (5-2)!}$$

and $6-2$ parts from $50-5$ standard parts remaining:
$$C_{N-K}^{n-k} = \frac{(N-K)!}{(n-k)! \cdot (N-K-n+k)!}\\
C_{45}^4 = \frac{45!}{4! \cdot (45-4)!}$$

Event $A$ will include all combinations of the previous two placements:

$$C(A) = C_K^k \cdot C_{N-K}^{n-k}$$

Thus, the sought probability is the ratio of the number of combinations that contribute to the event $A$ to the total number of combinations:

$$P(A) = \frac{C_K^k \cdot C_{N-K}^{n-k}}{C_N^n} = \frac{5! \cdot 45! \cdot 6! \cdot 44!}{2! \cdot 3! \cdot 4! \cdot 41! \cdot 50!} = \frac{4 \cdot 5 \cdot 5 \cdot 6 \cdot 42 \cdot 43 \cdot 44}{2 \cdot 46 \cdot 47 \cdot 48 \cdot 49 \cdot 50} = 0.094$$
or 9,4%.

### Task 3

Three dice are thrown. What is more likely: to get a total of 11 or 12 points?
Solve programmatically, generating all possible combinations.

---
Resolution:

```python
import itertools

def counter(dice_sides, num_dice):
    result = itertools.product(range(1, dice_sides+1), repeat=3)

    summary = {}
    for i in result:
        total = sum(i)
        if total in summary:
            summary[total] += 1
        else:
            summary[total] = 1

    print("| Points | Count |")
    for points, count in summary.items():
        print(f"|{points:^8}|{count:^7}|")

    return summary
```

```python
dice_sides = 6
num_dice = 3

result = counter(dice_sides, num_dice)

if result[11] > result[12]:
    print(f"It's more likely to get a total of 11 points: {result[11]} vs {result[12]} counts.")
elif result[12] > result[11]:
    print(f"It's more likely to get a total of 12 points: {result[12]} vs {result[11]} counts.")
else:
    print(f"Both totals have the same likelihood: {result[11]} counts each")
```

### Task 5

Two concentric circles are drawn on the plane (they have a common center),
the radii of which are 5 and 10 cm, respectively. Find the probability that a point thrown at random into a large circle will also fall into the ring formed by the constructed circles.

Solve analytically.

---
Resolution:

Area of a large circle:
$$ A_{10} = \pi R^2 = \pi \cdot 10^2 = 100 \pi $$
Ring area:
$$ A_{10-5} = \pi (R^2 - r^2) = \pi (10^2 - 5^2) = 75 \pi$$
Thus, the probability of hitting the ring when hitting a large circle will be the ratio of the areas of the ring and the large circle:
$$ P = \frac{A_{10-5}}{A_{10}} = \frac{75\pi}{100\pi} = \frac{3}{4} = 75\% $$

### Task 6

An urn contains one ball known to be either white or black with equal probabilities. A white ball is placed in the urn and then one ball is taken out at random. She turned out to be white. What is the probability that the remaining ball is white? Solve analytically.

---
Resolution:

Let's enter the following notation: \
$A$ is the event of drawing the first ball, which is white,\
$B$ – the only ball in the urn is white,\
then $\neg B$ - the only ball in the urn is black.\
According to the condition of the problem at the beginning of the experiment, the following two hypotheses have the same probability, therefore:\
$ H(B) = H(\neg B) = \frac{1}{2} $


Now let's calculate what was the probability of drawing the first white ball by entering the following symbols:\
$ P(A|H(B))$ is the probability of drawing the first ball, which is white, provided that a white ball was placed next to the white one, of course, it is equal to 1,\
$ P(A|H(\neg B))$ is the probability of drawing the first ball, which is white, provided that a white ball was added to the black one, it will be equal to $\frac{1}{2}$.\
Then the probability of drawing the white ball first according to the formula of total probability:
$$ P(A) = \sum_{i=1}^nP(H_i)P(A|H_i) = P(A|H(B)) \cdot H(B) + P(A|H(\neg B)) \cdot H(\neg B) = 1 \cdot \frac{1}{2} + \frac{1}{2} \cdot \frac{1}{2} = \frac{3}{4} $$

As a result of the observation, the event $A$ occurred, then the new conditional probability of the hypothesis that the remaining ball is white is calculated by the Bayes formula:
$$ P(H(B)| A) = \frac{P(H_i)P(A|H_i)}{\sum_{i=1}^nP(H_i)P(A|H_i)} = \frac{P(A|H(B)) \cdot H(B)}{P(A)} = \frac{1 \cdot \frac{1}{2}}{\frac{3}{4}} = \frac{2}{3}$$

## Task 7. Random wandering

The price of the company's shares at each moment of time can increase by 2 or decrease by 1 with equal probability. At the initial moment of time $t=0$, the price is equal to 0. Determine the average price of the share after 3 units of time ($t=3$). 
a) Solve analytically.

b) Write a simulation of this process. Calculate the average value and draw a histogram of the price for 10, 100, 1000 and 10000 simulations.

---
Resolution:

For the price after three time intervals, the following decision tree can be created, where the branch to the left is always a decrease in price, and the branch to the right is an increase:

t = 0...................................0\
......................................./......\\\
t = 1...........................-1.............2\
.............................../......\\......../......\\\
t = 2......................-2...........1...........4\
.........................../......\\..../.....\\..../.......\\\
t = 3.................-3 .........0..........3...........6
             

Thus, after three time slots, only four prices are possible, with -3 and 6 only being obtained in one way out of eight, and 0 and 3 in three ways out of eight.

The mathematical expectation, or average price, will be:
$$ M(p) = -3 \cdot \frac{1}{8} + 0 \cdot \frac{3}{8} + 3 \cdot \frac{3}{8} + 6 \cdot \frac{1}{8} = \frac {12}{8} = 1.5 $$

```python
import numpy as np


def stock_price_at_time(t: int) -> int:
    price = 0
    if t > 0:
        for i in range(t):
            if np.random.choice([0, 1]) == 1:
                price += 2
            else:
                price -= 1
    return price


def simulate_n_times(n: int, t: int) -> list:
    prices = []
    for i in range(n):
        np.random.seed()
        price = stock_price_at_time(t)
        prices.append(price)
    return prices
```

```python
import matplotlib.pyplot as plt

for n in [10, 100, 1000, 10000]:
    prices = simulate_n_times(n, 3)
    average_price = sum(prices) / len(prices)
    plt.hist(prices, bins=np.arange(min(prices), max(prices)+2)-0.5, density=True, alpha=0.75)
    plt.xlabel('Price')
    plt.ylabel('Probability')
    plt.title(f'Histogram of Share Prices after {n} simulation')
    plt.xticks(np.arange(min(prices), max(prices)+1))
    plt.axvline(average_price, color='r', linestyle='dashed', linewidth=1, label=f'Average Price: {average_price:.2f}')
    plt.grid(axis='y')
    plt.legend()
    plt.show()
```
