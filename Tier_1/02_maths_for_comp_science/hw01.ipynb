### Task 1
The template contains the numerical vectors $\overline{a}$ and $\overline{b}$. Calculate the following:
- the sum of $\overline{a}$ and $\overline{b}$
- the difference between $\overline{a}$ and $\overline{b}$
- the sum of $\overline{a}$ and $\overline{b}^T$. Explain the obtained result.
- matrix product (dot product) $\overline{a}$ and $\overline{b}^T$.
- matrix product (dot product) $\overline{a}$ and $\overline{b}$. Explain the obtained result.
- Hadamard product $\overline{a}$ and $\overline{b}$. Explain the obtained result.
- division of $\overline{a}$ and $\overline{b}$. Explain the obtained result.
- division of $\overline{a}$ and $\overline{b}^T$. Explain the obtained result.

```python
import numpy as np
a = np.array([[1, 2, 3, 4, 5]])
b = np.array([[1/2, 1, 2, 3, 4]])
print(a.shape)
print(b.shape)
```

#### 1.1 - Sum of $\overline{a}$ and $\overline{b}$

```python
res = a + b
res
```

#### 1.2 - Difference of $\overline{a}$ and $\overline{b}$

```python
res = a - b
res
```

#### 1.3 - Sum of $\overline{a}$ and $\overline{b}^T$. Explain the obtained result.

```python
# Explanation: Colab performed an addition operation between matrices of different sizes (1x5) and (5x1), adding elements from matrix b alternately to matrix a, thus forming a new matrix of size (5x5). In fact, it is not possible to add matrices of different sizes from a mathematical point of view, but it is valid from a programming point of view.
res = a + b.T
res
```

#### 1.4 - Matrix product (dot product) of $\overline{a}$ and $\overline{b}^T$.

```python
res = a @ b.T
res
```

#### 1.5 - Matrix product (dot product) of $\overline{a}$ and $\overline{b}$. Explain the obtained result.

```python
# Explanation: It is not possible to obtain a matrix product between two matrices in which the number of columns of the first matrix does not coincide with the number of rows of the second. In this case, we have the dimensions (1x5) and (1x5).
res = a @ b
res
```

#### 1.6 - Hadamard product of $\overline{a}$ and $\overline{b}$. Explain the obtained result.

```python
# Explanation: The Hadamard product is the usual element-by-element multiplication of matrices.
res = a * b
res
```

#### 1.7 - Division of $\overline{a}$ and $\overline{b}$. Explain the obtained result.

```python
# Explanation: The usual element-wise division of matrices
res = a / b
res
```

#### 1.8 - Division of $\overline{a}$ and $\overline{b}^T$. Explain the obtained result.

```python
# Explanation: It is not possible to divide matrices of different sizes even element by element. This result is not valid from a mathematical point of view, but it is valid from a programming point of view.
res = a / b.T
res
```

#### 2.1 - Reduce the $x$ vector by 2 times along the OX axis and increase it by 3 times along the OY axis

$$
M_1 =
\left (\begin{array}{cc}
1/2 & 0\\
0 & 3
\end{array}\right)
$$

$$
M_1x =
\left(\begin{array}{cc}
1/2 & 0\\
0 & 3
\end{array}\right)
\left (\begin{array}{cc}
2\\
1
\end{array}\right)=
\left(\begin{array}{cc}
1/2 \cdot 2 + 0 \cdot 1\\
0 \cdot 2 + 3 \cdot 1
\end{array}\right)=
\left(\begin{array}{cc}
1\\
3
\end{array}\right)
$$

```python
x = np.array([[2], [1]])
M1 = np.array([[1/2, 0], [0, 3]])
res = M1 @ x
res
```

$$

$$

```python
M2 = np.array([[-1, 0], [0, -1]])
res = M2 @ x
res
```

### Task 3

It is necessary to perform the same points as in Task 2, but applying the transformation to an arbitrary image from the Internet using the function cv.warpPerspective from the OpenCV module, which, in addition to image transformation tools, also provides tools for analyzing the content of images, in particular, recognizing objects in photos, tracking the movement of objects, applying machine learning methods and identifying common elements in different images.

All matrices must be set manually, and not through the get transformation functions.

```python
import cv2 as cv
import urllib
import numpy as np
from google.colab.patches import cv2_imshow as cv_imshow

# Read an image
def read_image_by_url(url):
    req = urllib.request.urlopen(url)
    arr = np.asarray(bytearray(req.read()), dtype=np.uint8)
    img = cv.imdecode(arr, -1)
    return img

url = 'https://i.pinimg.com/736x/65/cd/25/65cd25bd0833af5d3d7c5edf6573239b.jpg'

img = read_image_by_url(url)
```

```python
scale = np.float32([[0.5, 0, 0], [0, 0.5, 0], [0, 0, 1]])
img = cv.warpPerspective(img, scale, (int(img.shape[1]*0.5), int(img.shape[0]*0.5)))
```

```python
cv_imshow(img)
```

```python
img.shape
```

```python
rows, cols, rgb = img.shape
```

#### 3.1 - Image reduction by 2 times along the OX axis and 3 times along the OY axis

```python
M1 = np.float32([[0.5, 0, 0], [0, 3, 0], [0, 0, 1]])
res = cv.warpPerspective(img, M1, (int(cols*0.5), rows*3))
cv_imshow(res)
```

#### 3.2 - Display of the image relative to the origin of coordinates

```python
M2 = np.float32([[-1, 0, cols], [0, -1, rows], [0, 0, 1]])
res = cv.warpPerspective(img, M2, (cols, rows))
cv_imshow(res)
```

#### 3.3 - Moving the image by -30 on the $OX$ axis and by 10 on the $OY$ axis

```python
M3 = np.float32([[1, 0, -30], [0, 1, 10], [0, 0, 1]])
res = cv.warpPerspective(img, M3, (cols, rows))
cv_imshow(res)
```

#### 3.4 - Shifting the image by 60° along the $OY$ axis

```python
M4 = np.array([[1, 0, 0], [np.tan(np.deg2rad(60)), 1, 0], [0, 0, 1]])
res = cv.warpPerspective(img, M4, (cols, rows*3))
cv_imshow(res)
```

#### 3.5 - Rotate the image by 30°

```python
M5 = np.array([[np.cos(np.deg2rad(30)), -np.sin(np.deg2rad(30)), 0],
                [np.sin(np.deg2rad(30)), np.cos(np.deg2rad(30)), 0], [0, 0, 1]])
res = cv.warpPerspective(img, M5, (cols, rows))
cv_imshow(res)
```

#### 3.6 - Combining transformations from steps 1, 2, 4, 5 into one matrix and applying it to the image $x$

```python
M6 = M5 @ M4 @ M2 @ M1
res = cv.warpPerspective(img, M6, (cols, rows*3))
cv_imshow(res)
```
