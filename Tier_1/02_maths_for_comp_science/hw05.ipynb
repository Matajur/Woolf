### Task 1
Differentiate the function and find the value of the derivatives at the points $x = 1$ and $x=−1/2$. Solve programmatically and analytically.

```python
from sympy import diff, symbols
x = symbols("x")
x_1 = 1
x_2 = -1/2
```

#### 1.1 - Function
$f(x) = \frac{x^3}{3} + \frac{x^2}{2} - 2x$

$f'(x) = \frac{3x^{3-1}}{3} + \frac{2x^{2-1}}{2} - 1\cdot2x^{1-1} = x^2 + x - 2$

```python
f_1 = x ** 3 / 3 + x ** 2 / 2 - 2 * x
df_1 = diff((f_1))
df_1
```

$f'(x)|_{x=1} = 1^2 + 1 - 2 = 0$

```python
df_1.subs(x, x_1)
```

$f'(x)|_{x=-1/2} = \left(\frac{-1}{2}\right)^2 - \frac{1}{2} - 2 = \frac{1}{4} - \frac{1}{2} - 2 = -2\frac{1}{4}$

```python
df_1.subs(x, x_2)
```

#### 1.2 - Function
$f(x) = \sqrt{x^2 + 1}$

Let $u(x) = x^2 + 1$,\
then $f(u) = u^{\frac{1}{2}}$.\
The derivative of a complex function in this case:\
$f'(x) = \frac{df(u)}{du} \cdot \frac{du(x)}{dx} = \frac{1}{2}u^{\frac{1}{2}-1} \cdot 2x^{2-1} = \frac{1}{2\sqrt{u}} \cdot 2x = \frac{x}{\sqrt{x^2 + 1}}$

```python
f_2 = (x ** 2 + 1) ** (1/2)
df_2 = diff((f_2))
df_2
```

$f'(x)|_{x=1} = \frac{1}{\sqrt{1^2+1}} = \frac{\sqrt{2}}{2}$

```python
df_2.subs(x, x_1)
```

$f'(x)|_{x=-1/2} = \frac{-\frac{1}{2}}{\sqrt{\left(\frac{-1}{2}\right)^2+1}} = \frac{-\frac{1}{2}}{\sqrt{\frac{5}{4}}} = \frac{-1}{\sqrt{5}} = \frac{-\sqrt{5}}{5}$

```python
df_2.subs(x, x_2)
```

#### 1.3 - Function
$f(x) = \frac{1}{\sqrt{x^2 + 1}}$

Let $u(x) = x^2 + 1$,\
then $f(u) = u^{-\frac{1}{2}}$.\
The derivative of a complex function in this case:\
$f'(x) = \frac{df(u)}{du} \cdot \frac{du(x)}{dx} = -\frac{1}{2}u^{-\frac{1}{2}-1} \cdot 2x^{2-1} = \frac{-1}{2\sqrt{u^3}} \cdot 2x = \frac{-x}{\sqrt{(x^2 + 1)^3}}$

```python
f_3 = (x ** 2 + 1) ** (-1/2)
df_3 = diff((f_3))
df_3
```

$f'(x)|_{x=1} = \frac{-1}{\sqrt{(1^2+1)^3}} = \frac{-1}{2\sqrt{2}} = \frac{-\sqrt{2}}{4}$

```python
df_3.subs(x, x_1)
```

$f'(x)|_{x=-1/2} = \frac{-\left(-\frac{1}{2}\right)}{\sqrt{\left( \left(\frac{-1}{2}\right)^2+1 \right)^3}} = \frac{\frac{1}{2}}{\sqrt{\left( \frac{5}{4}\right)^3}} = \frac{1}{\sqrt{5}} \cdot \frac{4}{5} = \frac{4\sqrt{5}}{25}$

```python
df_3.subs(x, x_2)
```

### Task 2
Differentiate the function analytically and find the maximum point. Programmatically visualize the function, its derivative and maximum point.\
$y = -3x^2 + 30x$

#### 2.1 - Analytical solution

Похідна:\
$\frac{dy}{dx} = -3 \cdot 2x^{2-1} + 1 \cdot 30x^{1-1} = -6x + 30$

Екстремум:\
$\frac{dy}{dx} = -6x + 30 = 0$\
$x = \frac{-30}{-6} = 5$\
$y|_{x=5} = -3 \cdot 5^2 + 30 \cdot 5 = 5 \cdot (30 - 15) = 75$

#### 2.2 - Program visualization

```python
import matplotlib.pyplot as plt
from sympy import diff, symbols, solve

fig, ax = plt.subplots()

# Move the left and bottom columns to x = 0 and y = 0 respectively
ax.spines[["left", "bottom"]].set_position(("data", 0))

# Hide the top and right lines
ax.spines[["top", "right"]].set_visible(False)

# Draw arrows (as black triangles: ">k"/"^k") at the ends of the axes
# Also disable clipping (clip_on=False) of arrows
ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False)

# Enable grid
ax.grid(True, linestyle='-.')

# Set the function
x, y = symbols('x y')
fx = -3 * x ** 2 + 30 * x

# Calculate the derivative of the function
dx = diff((fx))
k = []
n = []
l = []

for i in range(15):
    j = i - 2
    k.append(j)
    n.append(dx.subs(x, j))
    l.append(fx.subs(x, j))

# Extremum point
x_extr = solve(dx)[0]
y_extr = fx.subs(x, x_extr)

# Draw the graph
ax.plot(k, n, label="derivative")
ax.plot(k, l, label="function")

x_coords = [x_extr, x_extr]
y_coords = [0, y_extr]
plt.plot(x_coords, y_coords, linestyle='--', color="g")

x_coords = [0, x_extr]
y_coords = [y_extr, y_extr]
plt.plot(x_coords, y_coords, linestyle='--', color="g")

plt.xlabel('X', loc="right", fontweight='bold')
plt.ylabel('Y', loc="top", fontweight='bold')
plt.title('Graph of the parabolic function, its derivative and extremum point', fontweight='bold')
description = f'max ({round(x_extr, 1)}, {round(y_extr, 1)})'
plt.text(x_extr, y_extr, description, fontsize=10, ha='right', va='bottom')
plt.legend(loc='best')

plt.show()
```
